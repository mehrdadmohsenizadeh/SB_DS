**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data?

You might need to **melt** your DataFrame here.

```python
# retain "London_Borough" and "ID" columns as identifier variables after
# melting DataFrame "properties_T".

# NOTE: All other columns will be "melted" into two new columns:
#       1. variable names (column "0")
#       2. corresponding values (column "value")
clean_properties = pd.melt(
      properties_T
    , id_vars = ['London_Borough', 'ID']
)

clean_properties
```

```python
# rename columns '0' and 'value' to
# 'Month' and 'Average_price', respectively.
clean_properties = clean_properties.rename(
    columns = {
          0       : 'Month'
        , 'value' : 'Average_price'
      }
)

clean_properties.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers...

```python
type(clean_properties['Average_price'])
```

```python
# convert the values type of column 'Average_price' from a 'Series' to numeric type.
clean_properties['Average_price'] = pd.to_numeric(
    clean_properties['Average_price']
)

clean_properties
```

```python
clean_properties.dtypes
```

> As shown, the data type of column `Average_price` was successfully converted from Panda's `Series` to `float64`.

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the `ID`, `Average_price`, `Month`, and `London_Borough` columns? Remember that there are only $32$ London Boroughs. How many entries do you have in that column?

Check out the contents of the London Borough column, and if you find null (`NaN`) values, get rid of them however you see fit.

```python
# Unique values (Boroughs) in column 'London_Borough'
print(f"Number of unique London boroughs and non-borughs: {clean_properties['London_Borough'].nunique()}")
print(f"Number of entries: {clean_properties['London_Borough'].count()}")
```

```python
# in dataframe 'clean_properties', retain only those rows whose corresponding
# 'Average_price' value is not NaN.
NaNFreeDF1 = clean_properties[clean_properties['Average_price'].notna()]
NaNFreeDF1
```

> Within the dataframe `clean_properties`, which comprises a total of $16,368$ entries, a significant majority of $15,345$ entries possess no `NaN` value within their corresponding `Average_price` column.

> Now, let's drop any row which contains a `NaN` value in dataframe `clean_properties` and call the new dataframe as `NaNFreeDF2`.

```python
NaNFreeDF2 = clean_properties.dropna()
NaNFreeDF2
```

> Surprisingly, we've obtained the same outcome, suggesting that the dataframe `clean_properties` is devoid of `NaN` values in its remaining columns as well (i.e., both `NaNFreeDF1` and `NaNFreeDF2` have the same shape, $15345$ $\text{rows}$ $\times$ $4$ $\text{columns}$).

```python
# drop all non-boroughs from column "London_Borough".
nonBoroughs = [
      'Inner London'
    , 'Outer London'
    , 'NORTH EAST'
    , 'NORTH WEST'
    , 'YORKS & THE HUMBER'
    , 'EAST MIDLANDS'
    , 'WEST MIDLANDS'
    , 'EAST OF ENGLAND'
    , 'LONDON'
    , 'SOUTH EAST'
    , 'SOUTH WEST'
    , 'England'
]

NaNFreeDF2 = NaNFreeDF2[~NaNFreeDF2.London_Borough.isin(nonBoroughs)]
NaNFreeDF2
```

```python
print(f"Number of unique London boroughs after filtering non-boroughs: {NaNFreeDF2['London_Borough'].nunique()}")
```

> After dropping all non-boroughs from column `London_Borough` we ended up having $11,253$ columns in the dataframe `NaNFreeDF2`. Remember that dataframe `NaNFreeDF2` was achieved by dropping all `NaN` values from the original dataframe `clean_properties`.

```python
# create a copy of dataframe 'NaNFreeDF2' and call it 'df'.
df = NaNFreeDF2
df.head()
```

**3. Modeling**

Consider creating a function called `create_price_ratio` that will calculate a <b>ratio of house prices</b>, comparing the price of a house in <b>2018</b> to the price in <b>1998</b>. Call this function as follows:

1. Take a filter of `dfg`, specifically where this filter constrains the `London_Borough` column, as an argument. For example, one admissible argument should be: `dfg[dfg['London_Borough']=='Camden']`.
2. Get the `Average_price` for that Borough, for the years <b>1998</b> and <b>2018</b>.
4. Calculate the <b>ratio</b> of the `Average_price` for <b>1998</b> divided by the `Average_price` for <b>2018</b>.
5. Return that <b>ratio</b>.

Once you've written this function, you ultimately want to use it to iterate through all the unique `London_Boroughs` and work out the <b>ratio</b> capturing the difference of house prices between <b>1998</b> and <b>2018</b>.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great!

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(dataframe):
    y1998 = float(dataframe['Average_price'][dataframe['Year']==1998])
    y2018 = float(dataframe['Average_price'][dataframe['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
create_price_ratio(dfg[dfg['London_Borough'] == 'Barking & Dagenham'])
```

```python
final = {}

for london_borough in dfg['London_Borough'].unique():
    borough  = dfg[dfg['London_Borough']== london_borough]
    final[london_borough] = create_price_ratio(borough)

print(final)
```

```python
df_ratios = pd.DataFrame(final)
df_ratios.head()
```

```python
df_ratios_T = df_ratios.T
df_ratios   = df_ratios_T.reset_index()
df_ratios.head()
```

```python
df_ratios.rename(
      columns = {
            'index' :'Borough'
          , 0       :'2018'
      }
    , inplace = True
)

df_ratios
```

```python
df_ratios_top_15 = df_ratios.sort_values(by='2018', ascending = False).head(15)
df_ratios_top_15
```

```python
ax = df_ratios_top_15[['Borough', '2018']].plot(kind='bar')
ax.set_xticklabels(df_ratios_top_15.Borough)
```

```python
dfg.max()
```

```python
df_ratios_top_15_table = pd.melt(
      df_ratios_top_15
    , id_vars = ['Borough']
)

df_ratios_top_15_table
```

```python
def plot_average_prices(input_df, boroughs):

    plt.figure(figsize=(7, 4.2))

    for borough in boroughs:
        df_borough = input_df[input_df['London_Borough'] == borough]
        sns.lineplot(
              x     ='Month'
            , y     ='Average_price'
            , data  = df_borough
            , label = borough
        )

    plt.xlabel('Year')
    plt.ylabel('Average Price')
    plt.title('Average Monthly London House Price by Borough (top 5)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

```python
top_5_boroughs = df_ratios_top_15_table['Borough'][:5]

plot_average_prices(
      df
    , top_5_boroughs
)
```
